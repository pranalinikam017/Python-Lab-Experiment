```python
import math

def calculate_area():
    print("--- Shape Area Calculator ---")
    print("1. Circle")
    print("2. Rectangle")
    print("3. Triangle")
    
    choice = input("\nEnter the number of the shape (1-3): ")

    if choice == '1':
        radius = float(input("Enter the radius: "))
        area = math.pi * (radius ** 2)
        print(f"\nThe area of the circle is: {area:.2f}")

    elif choice == '2':
        length = float(input("Enter the length: "))
        width = float(input("Enter the width: "))
        area = length * width
        print(f"\nThe area of the rectangle is: {area:.2f}")

    elif choice == '3':
        base = float(input("Enter the base: "))
        height = float(input("Enter the height: "))
        area = 0.5 * base * height
        print(f"\nThe area of the triangle is: {area:.2f}")

    else:
        print("Invalid selection. Please run the program again.")

if __name__ == "__main__":
    calculate_area()
```

#### Single Inheritance

```python
class Parent:
    def __init__(self):
        self.name = "JK"
        self.age = 28
        self.home = "Mauli"

class Child(Parent):   # Inheriting from Parent
    def __init__(self):
        super().__init__()   # Call Parent constructor
        self.car = "Toyota"
        self.mobileno = 1234567890

# Create object of Child
c = Child()

# Access Parent + Child properties
print("Name of Parent :", c.name)
print("Age of Parent :", c.age)
print("Home Name :", c.home)
print("Car :", c.car)
print("Mobile No :", c.mobileno)
```

##### Multilevel Inheritance

```python
class Grandparent:
    def __init__(self):
        self.name = "Ramesh"
        self.farm = "2 acres"

class Parent(Grandparent):
    def __init__(self):
        super().__init__()
        
        self.age = 28
        self.home = "Mauli"

class Child(Parent):   # Inheriting from Parent
    def __init__(self):
        super().__init__()   # Call Parent constructor
        self.car = "Toyota"
        self.mobileno = 1234567890

# Create object of Child
c = Child()

# Access Parent + Child properties
print("Name of GrandParent :", c.name)
print("Farm : ",c.farm)
print("Age of Parent :", c.age)
print("Home Name :", c.home)
print("Car :", c.car)
print("Mobile No :", c.mobileno)
```

#### Multiple

```python
class Father:
    def __init__(self):
        self.home = "Mauli"
        self.farm = "2 acres"

class Mother:
    def __init__(self):
        self.name = "jacqline"
        self.age = 28

class Child(Father, Mother):  # Inheriting from Father and Mother
    def __init__(self):
        # We call both parent constructors explicitly 
        # so the child gets all attributes.
        Father.__init__(self)
        Mother.__init__(self)
        
        self.car = "Toyota"
        self.mobileno = 1234567890

# Create object of Child
c = Child()

# Output the results
print("--- Family Properties ---")
print("Name of home   :", c.home)
print("Farm           :", c.farm)
print("Name of mother :", c.name)
print("Age of mother  :", c.age)
print("Car of child   :", c.car)
print("Mobile no      :", c.mobileno)
```

#### Polymorphism

```python
class Calculator:
    def multiply(self, a= 1, b=1, *args):
        result = a*b
        for num in args:
            result *= num
        return result
    
calc = Calculator()

print(calc.multiply())
print(calc.multiply(4))

print(calc.multiply(2,3))
print(calc.multiply(2, 3, 4,))
```

#### Method Overriding

```python
class Animal:
      def speak(self):
            print("The animal makes a sound")

class Dog(Animal):
      def speak(self):
            print("The dog barks.")

class Cat(Animal):
      def speak(self):
            super().speak()
            print("The cat meows")


gc = Animal()
puppy = Dog()
kit = Cat()

print("Generic Animal: ")
gc.speak()
print("\nDog : ")
puppy.speak()
print("\nCat : ")
kit.speak()
```

#### Encapsulation

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name 
        self.salary = salary
emp = Employee("jack", 50000)
print(emp.name)
print(emp.salary)
```
