```python
st1 = {"Apple","Cherry", "Strawberry"}

print("Set1 : ",st1)

st1.add("Grapes")
print("After add Operation : ",st1)

st1.remove("Cherry")
print("After remove operation :",st1)

st2 = {10, 20, 50, 60}
print("Set 2 :",st2)

u = st1.union(st2)
print("Union of set1 and set2 is :",u)

st1.clear()
print("Using clear method on set1: ",st1)

st3 ={23, 67, 50, 80}
print("Set 3 is : ",st3)

s = st2.intersection(st3)

print("Intersection of set2 and set3: ",s)

p = st2.difference(st3)
print("Difference between set2 and set3 : ",p)
```

```python
t1 = []

n = int(input("How many numbers you want to add in tuple:"))

for i in range(n):
    num = int(input("Enter a number : "))
    t1.append(num)
t1 = tuple(t1)

total = sum(t1)

avg = total / n
print("Total sum : ", total)
print("Average of Number: ", avg)
```
