```python
num = int(input("Enter number:"))

copyN = num
lastdigit = 0
revNum = 0
while(num > 0):
    lastdigit = num % 10
    revNum = (revNum*10) + lastdigit
    num = num // 10

if(copyN == revNum):
    print("Number is Palindrome number")
else:
    print("Number is not Palindrome number")
```

```python
# Program to find GCD and LCM of two numbers

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

a = num1
b = num2

# Finding GCD using Euclidean Algorithm
while b != 0:
    a, b = b, a % b

gcd = a

# Finding LCM
lcm = (num1 * num2) // gcd

print("GCD =", gcd)
print("LCM =", lcm)
```
