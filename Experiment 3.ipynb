##### Is palindrome or not

```python
str1 = input("Enter a string: ")
 


if(str1 == str1[::-1]):
    print("String is palindrome")
else:
    print("String is not palindrome")
```

#### Is Anagram

```python
str1 = input("Enter string 1: ")
str2 = input("Enter string 2: ")

if sorted(str1) == sorted(str2):
    print("The two Strings are Anagrams")
else:
    print("The two strings are not Anagrams")
```

#### Frequency

```python
text = "programming"
frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequencies:", frequency)
```

#### Count Vowels and Consonants

```python
str1 = input("Enter a string: ")

countVowel = 0
countcon = 0

for ch in str1:
    if ch in "aeiouAEIOU":
        countVowel += 1
    elif ch.isalpha():
        countcon += 1

print("Count of vowels in string is:", countVowel)
print("Count of consonants is:", countcon)
```

#### Convert to uppercase And Lowercase

```python
text = "Hello, Python 2026!"

# Convert to Uppercase
upper_text = text.upper()

# Convert to Lowercase
lower_text = text.lower()

# Convert to Title Case (First letter of each word capitalized)
title_text = text.title()

# Convert to Swap Case (Inverts upper to lower and vice versa)
swap_text = text.swapcase()

print(f"Original:   {text}")
print(f"Uppercase:  {upper_text}")
print(f"Lowercase:  {lower_text}")
print(f"Title Case: {title_text}")
print(f"Swap Case:  {swap_text}")
```

#### Replace a string

```python
original_text = "I love apples. Apples are the best fruit."

new_text = original_text.replace("apples", "oranges")

print(f"Original: {original_text}")
print(f"Updated:  {new_text}")
```
