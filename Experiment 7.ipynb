```python
def evenOdd(n):
    if( n % 2 == 0):
        print("Even Number")
    else :
        print("Odd Number")

evenOdd(18)
evenOdd(5)

```

##### Return multiple values from a function

```python
def rect(l, w):
    area = l * w
    perimeter = 2 * (l + w)
    return area, perimeter

my_area, my_per = rect(5, 8)

print("The area of Rectangle is : ", my_area)
print("The perimeter of Rectangle is : ", my_per)
    
```

##### Factorial

```python
def fact(n):
    if(n == 0):
        return 1
    else :
        result = 1
        for i in range(1, n+1):
              result =  result * i
        return result
    
n = 6
    
print("Factorial of ",n ,"is : ", fact(6))
```

#### Fibonacci series

```python
def fib(n):
    if(n == 0):
       return 0
    elif(n == 1):
        return 1
    else:
        return fib(n-1) + fib(n-2)

terms = 7
print("Fibonacci Series:")
for i in range(0, terms+1):
       print(fib(i))
```

#### Is Prime Function

```python
def is_prime(n):
    if n < 2:
        return "Not a prime number"
    
    # Assume it is prime until proven otherwise
    for i in range(2, n): # We only check up to n-1
        if n % i == 0:
            # We found a divisor! It's definitely NOT prime.
            return "Not a prime number"
            
    # If the loop finishes without finding a divisor, it IS prime.
    return "Prime Number"

print(is_prime(9))
```

#### Recursive Function

```python
def fact(n):
    if(n == 0):
        return 1
    else :
        return n * fact(n-1)
    
n = 6
    
print("Factorial of ",n ,"is : ", fact(6))
```

#### Using Built in function

```python
# A simple function to check health status
def health_check(temperature):
    if temperature > 38:
        return "Fever detected. Please consult a doctor."
    elif temperature < 36:
        return "Temperature is low. Stay warm."
    else:
        return "Temperature is normal."

# 1. Get user input
user_temp = float(input("Enter body temperature in Celsius: "))

# 2. Get the result from our function
status = health_check(user_temp)

# 3. Print the result
print(f"Assistant Result: {status}")
```
