```python
list1 = [12, 56, 34, 767, 89, 213, 908]

evenNum = 0
oddNum = 0
for i in list1:
    if(i % 2 == 0):
        evenNum = evenNum+1
    else:
        oddNum = oddNum+1

print("Count of even number in list1 : ",evenNum)
print("Count of odd number in list1 : ",oddNum)
```

```python
numbers = [10, 20, 30, 40, 50, 60]

print("Original List:", numbers)

# Slice elements from index 1 to 4
print("Elements from index 1 to 4:", numbers[1:5])

# Slice elements from beginning to index 3
print("First three elements:", numbers[:3])

# Slice elements from index 3 to the end
print("Elements from index 3 to end:", numbers[3:])

# Slice with step value
print("Every second element:", numbers[::2])

# Reverse the list using slicing
print("Reversed List:", numbers[::-1])
```

```python
numbers = []

# Take input from the user
n = int(input("How many numbers do you want to enter? "))

for i in range(n):
    num = float(input("Enter a number: "))
    numbers.append(num)

# Calculate sum

total = sum(numbers)

# Calculate average
average = total / n

# Display results
print("Numbers:", numbers)
print("Sum =", total)
print("Average =", average)
```

```python
t1 = []

n = int(input("How many numbers you want to add: "))

for i in range(n):
    num = float(input("Enter a number : "))
    t1.append(num)

# Convert the list to a tuple (as per your prompt's intent)
t1 = tuple(t1)

# Pass the entire collection (t1) to the sum function
total = sum(t1)

avg = total / n

print("Tuple content:", t1)
print("Total sum : ", total)
print("Average of Number: ", avg)
```
